#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
# For the date early in the dataset, measuring the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = coinbase['Close'].loc['2018-01-17 00:00:00' : '2018-01-17 08:00:00'] - bitstamp['Close'].loc['2018-01-17 00:00:00' : '2018-01-17 08:00:00']
# Using a conditional statement to generate the summary statistics for the arbitrage_spread DataFrame, arbitrage_spread_early
arbitrage_spread_early_sumstat = arbitrage_spread_early[arbitrage_spread_early > 0]
arbitrage_spread_early_sumstat.describe()
```

```python
# Using a conditional statement to generate the summary statistics for the arbitrage_spread DataFrame, arbitrage_spread_middle
arbitrage_spread_middle_sumstat = arbitrage_spread_middle[arbitrage_spread_middle > 0]
arbitrage_spread_middle_sumstat.describe()
```

```python
# Using a conditional statement to generate the summary statistics for the arbitrage_spread DataFrame, arbitrage_spread_late
arbitrage_spread_late_sumstat = arbitrage_spread_late[arbitrage_spread_late > 0]
arbitrage_spread_late_sumstat.describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculating the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early= arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp['Close'].loc['2018-01-17 00:00:00' : '2018-01-17 08:00:00']

# Reviewing the spread return DataFrame
spread_return_early.dropna()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early = spread_return_early[spread_return_early > .01]

# Reviewing the first five profitable trades using the head function
profitable_trades_early.head()
```

#### 4. Generate the summary statistics of your spread returns that are greater than 1%. How do the average returns compare among the three dates?

```python
# For the date early in the dataset, generatinf the summary statistics for the profitable trades
# or you trades where the spread returns are are greater than 1%
profitable_trades_early.describe()
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the date early in the dataset, calculating the potential profit per trade in dollars 
# Multiplying the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp['Close'].loc['2018-01-17 00:00:00' : '2018-01-17 08:00:00']

# Dropping any missing values from the profit DataFrame using the dropna function
profit_per_trade_early = profit_early.dropna()

# Viewing the early profit DataFrame and using the dropna function to remove NAN values
profit_early.dropna()
```

#### 6. Generate the summary statistics, and plot the results for each of the three DataFrames.

```python
# Generating the summary statistics for the early profit per trade DataFrame
profit_per_trade_early.describe()
```

```python
# Plotting the results for the early profit per trade DataFrame
profit_per_trade_early.plot(figsize=(20,8), title="Profit Per Trade for Bitstamp in the Early Hours of January 17, 2018", color="blue")
```

#### 7. Calculate the potential arbitrage profits that you can make on each day. To do so, sum the elements in the profit_per_trade DataFrame.

```python
# Calculating the sum of the potential profits for the early profit per trade DataFrame
profit_sum = profit_per_trade_early.sum()
profit_sum
```

#### 8. Using the `cumsum` function, plot the cumulative sum of each of the three DataFrames. Can you identify any patterns or trends in the profits across the three time periods?

```python
# Using the cumsum function to calculate the cumulative profits over time for the early profit per trade DataFrame
cumulative_profit_early = profit_per_trade_early.cumsum()
```

```python
# Plotting the cumulative sum of profits for the early profit per trade DataFrame
cumulative_profit_early.plot(figsize=(20,8), title="Cumulative Coinbase Profits")
```
